```python
# %% [markdown]
# # 武蔵野大学Webサイト サイトマップ抽出

# %%
# ライブラリを読み込む
import requests
from bs4 import BeautifulSoup
import time

# %%
# 最初のページにアクセス
url = "https://www.musashino-u.ac.jp/"
res = requests.get(url)
res.encoding = res.apparent_encoding
print(f"ステータスコード:{res.status_code}")

# %%
# 結果を保存する辞書
sitemap = {}

# まだ訪問していないURLのリスト
not_visited = ["https://www.musashino-u.ac.jp/"]

# もう訪問したURLのリスト
already_visited = []

# %%
# すべてのページを訪問する
count = 0
while len(not_visited) > 0:
    # リストの最初のURLを取り出す
    current_url = not_visited.pop(0)
    
    # もう訪問済みならスキップ
    if current_url in already_visited:
        continue
    
    # 訪問済みリストに追加
    already_visited.append(current_url)
    count = count + 1
    
    print(f"[{count}] {current_url}")
    
    try:
        # ページにアクセス
        res = requests.get(current_url, timeout=10)
        res.encoding = res.apparent_encoding
        soup = BeautifulSoup(res.content, 'html.parser')
        
        # タイトルを取得して辞書に保存
        title_tag = soup.find('title')
        if title_tag:
            sitemap[current_url] = title_tag.get_text()
        else:
            sitemap[current_url] = ""
        
        # ページ内のすべてのリンクを探す
        links = soup.find_all('a', href=True)
        for link in links:
            link_url = link['href']
            
            # 相対URLを絶対URLに変換
            if link_url.startswith('/'):
                link_url = "https://www.musashino-u.ac.jp" + link_url
            elif not link_url.startswith('http'):
                continue
            
            # #を削除
            link_url = link_url.split('#')[0]
            
            # 同じドメインで、まだ訪問していないなら追加
            if "musashino-u.ac.jp" in link_url:
                if link_url not in already_visited:
                    if link_url not in not_visited:
                        not_visited.append(link_url)
        
        # 1秒待つ（サーバーに負荷をかけないため）
        time.sleep(1)
        
    except:
        print(f"  → エラーが発生しました")

# %%
# 結果を表示
print("\n=== 結果 ===")
print(sitemap)
```

```python
# %% [markdown]
# # 武蔵野大学Webサイト サイトマップ抽出（初心者向け簡単版）

# %%
# 必要なライブラリを読み込む
import requests
from bs4 import BeautifulSoup, Comment
import time

# %%
# 最初のページにアクセス
url = "https://www.musashino-u.ac.jp/"
res = requests.get(url)
print(f"ステータスコード:{res.status_code}")

# %%
# スキップしたいファイルの拡張子
skip_files = ['.pdf', '.jpg', '.png', '.gif', '.zip', '.xlsx', '.docx']

# 結果を保存する場所
sitemap = {}

# これから訪問するURLのリスト
to_visit = ["https://www.musashino-u.ac.jp/"]

# すでに訪問したURLのリスト
visited = []

# %%
# ページを1つずつ訪問する
page_count = 0

while len(to_visit) > 0:
    # リストの最初のURLを取り出す
    current = to_visit[0]
    to_visit.remove(current)
    
    # すでに訪問していたらスキップ
    if current in visited:
        continue
    
    # ファイル（画像やPDF）かチェック
    is_file = False
    for ext in skip_files:
        if current.endswith(ext):
            is_file = True
            print(f"[スキップ] {current}")
            break
    
    if is_file:
        continue
    
    # 訪問済みリストに追加
    visited.append(current)
    page_count = page_count + 1
    print(f"[{page_count}] {current}")
    
    # ページを取得
    try:
        res = requests.get(current, timeout=10)
        soup = BeautifulSoup(res.content, 'html.parser')
        
        # HTMLコメントを削除
        for comment in soup.find_all(string=lambda text: isinstance(text, type(soup))):
            comment.extract()
        
        # ページのタイトルを取得
        title = soup.find('title')
        if title:
            sitemap[current] = title.get_text()
        else:
            sitemap[current] = "タイトルなし"
        
        # ページ内のリンクをすべて探す
        all_links = soup.find_all('a', href=True)
        
        for link in all_links:
            href = link['href']
            
            # 相対URLを絶対URLに変換
            if href.startswith('/'):
                href = "https://www.musashino-u.ac.jp" + href
            
            # httpで始まらないリンクはスキップ
            if not href.startswith('http'):
                continue
            
            # #以降を削除
            if '#' in href:
                href = href.split('#')[0]
            
            # ファイルかチェック
            is_file = False
            for ext in skip_files:
                if href.endswith(ext):
                    is_file = True
                    break
            
            if is_file:
                continue
            
            # 武蔵野大学のURLで、まだリストにないなら追加
            if "musashino-u.ac.jp" in href:
                if href not in visited and href not in to_visit:
                    to_visit.append(href)
        
        # 1秒待つ（サーバーに優しく）
        time.sleep(0.75)
        
    except:
        print(f"  エラー: このページは読み込めませんでした")

# %%
# 結果を表示
print("\n========== 結果 ==========")
print(f"訪問したページ数: {page_count}")
print(sitemap)
```
